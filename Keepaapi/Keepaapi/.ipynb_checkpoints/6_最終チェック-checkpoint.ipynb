```python
import requests
import time
import pandas as pd
import tkinter as tk
from tkinter import messagebox, filedialog
from tkinterdnd2 import DND_FILES, TkinterDnD
import threading
import datetime
import traceback
import ctypes
import os

# ============================================================
# è¨­å®šå€¤
# ============================================================
DOMAIN_JP = 5
STOP_FLAG = False
MAX_TOKENS_PER_ITEM = 10
MAX_SECONDS_ALLOWED = 10       # ã‚¿ã‚¤ãƒ ã‚¢ã‚¦ãƒˆï¼š10ç§’
ERROR_WAIT_TIME = 300          # ã‚¨ãƒ©ãƒ¼æ™‚ã®å¾…æ©Ÿæ™‚é–“ï¼ˆç§’ï¼‰ï¼5åˆ†
SAVE_INTERVAL = 10             # âœ… 10ä»¶ã”ã¨ã«ä¿å­˜

# ============================================================
# ã‚¹ãƒªãƒ¼ãƒ—é˜²æ­¢ï¼ˆWindowsï¼‰
# ============================================================
ES_CONTINUOUS = 0x80000000
ES_SYSTEM_REQUIRED = 0x00000001

def prevent_sleep():
    try:
        ctypes.windll.kernel32.SetThreadExecutionState(ES_CONTINUOUS | ES_SYSTEM_REQUIRED)
    except Exception:
        pass

def allow_sleep():
    try:
        ctypes.windll.kernel32.SetThreadExecutionState(ES_CONTINUOUS)
    except Exception:
        pass

# ============================================================
# Keepa APIå‘¼ã³å‡ºã—
# ============================================================
def fetch_top_display_price(api_key: str, code: str):
    url = (
        f"https://api.keepa.com/product?key={api_key}"
        f"&domain={DOMAIN_JP}&code={code}"
        "&history=0&offers=20&onlyLiveOffers=0&buybox=1&stats=0"
    )
    start_time = time.time()
    try:
        resp = requests.get(url, timeout=MAX_SECONDS_ALLOWED)
        if resp.status_code == 429:
            return None, None, "ãƒˆãƒ¼ã‚¯ãƒ³æ¯æ¸‡", 0
        data = resp.json()
    except Exception as e:
        return None, None, f"é€šä¿¡ã‚¨ãƒ©ãƒ¼: {e}", 0

    if time.time() - start_time > MAX_SECONDS_ALLOWED:
        return None, None, f"å‡¦ç†æ™‚é–“è¶…éï¼ˆ{MAX_SECONDS_ALLOWED}ç§’ï¼‰", 0
    if not data or "products" not in data:
        return None, None, "ãƒ‡ãƒ¼ã‚¿ãªã—", 0

    products = data["products"]
    if not products:
        return None, None, "å•†å“ãŒè¦‹ã¤ã‹ã‚‰ãªã„", 0

    product = products[0]
    title = product.get("title", "")
    stats = product.get("stats") or {}

    # âœ… BuyBoxå„ªå…ˆ
    for key in ("buyBoxPrice", "buyBoxShippingPrice", "current_BUY_BOX_SHIPPING"):
        v = stats.get(key)
        if isinstance(v, (int, float)) and v > 0:
            return title, int(v), None, 0

    # âœ… Primeå„ªå…ˆï¼ˆãªã‘ã‚Œã°æœ€åˆã®å‡ºå“ï¼‰
    offers = product.get("offers") or []
    order = product.get("liveOffersOrder") or []
    ordered = [offers[i] for i in order if isinstance(i, int) and i < len(offers)]
    if not ordered and offers:
        ordered = offers

    prime_offer = next((o for o in ordered if o.get("isPrime")), None)
    chosen = prime_offer or (ordered[0] if ordered else None)

    if chosen:
        price = chosen.get("price")
        ship = chosen.get("shipping") or 0
        if price and price > 0:
            total = int(price) + int(ship)
            return title, total, None, 0

    hit_count = len(offers)
    if hit_count > 0:
        return title, None, f"ä¾¡æ ¼å–å¾—å¤±æ•—ï¼ˆ{hit_count}ä»¶ãƒ’ãƒƒãƒˆï¼‰", 0
    else:
        return title, None, "å•†å“ãŒè¦‹ã¤ã‹ã‚‰ãªã„", 0

# ============================================================
# ãƒ­ã‚°å‡ºåŠ›ã¾ã¨ã‚
# ============================================================
def flush_logs(log_box, buffer):
    """ãƒãƒƒãƒ•ã‚¡å†…ã®ãƒ­ã‚°ã‚’ç”»é¢ã«å³å‡ºåŠ›"""
    if not buffer:
        return
    log_box.insert(tk.END, "".join(buffer))
    log_box.see(tk.END)
    buffer.clear()

# ============================================================
# ãƒ¡ã‚¤ãƒ³å‡¦ç†
# ============================================================
def start_process(api_key, filepath, log_box, start_button):
    global STOP_FLAG
    STOP_FLAG = False
    prevent_sleep()
    start_button.config(state="disabled")

    try:
        df = pd.read_excel(filepath, header=None)
    except Exception as e:
        messagebox.showerror("èª­è¾¼ã‚¨ãƒ©ãƒ¼", f"Excelãƒ•ã‚¡ã‚¤ãƒ«ã‚’é–‹ã‘ã¾ã›ã‚“ã§ã—ãŸã€‚\n{e}")
        start_button.config(state="normal")
        allow_sleep()
        return

    total = len(df)
    timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
    desktop_path = os.path.join(os.path.expanduser("~"), "Desktop")
    output_file = os.path.join(desktop_path, f"çµæœ_{timestamp}.xlsx")

    log_box.insert(tk.END, f"ğŸ“˜ ãƒ•ã‚¡ã‚¤ãƒ«èª­è¾¼å®Œäº†: {filepath}\nğŸ”¢ å…¨{total}ä»¶ã®å‡¦ç†ã‚’é–‹å§‹ã—ã¾ã™ã€‚\n\n")
    log_box.see(tk.END)

    results = []
    log_buffer = []

    try:
        for i, row in df.iterrows():
            if STOP_FLAG:
                log_box.insert(tk.END, "ğŸ›‘ å¼·åˆ¶åœæ­¢ã‚’æ¤œå‡º â†’ ç¾åœ¨ã®çµæœã‚’ä¿å­˜ä¸­...\n")
                pd.DataFrame(results).to_excel(output_file, index=False)
                log_box.insert(tk.END, f"ğŸ’¾ ä¸­æ–­æ™‚ã®çµæœã‚’ä¿å­˜ã—ã¾ã—ãŸ â†’ {output_file}\n")
                break

            jan = str(row.iloc[0]).strip() if len(row) > 0 else ""
            if not jan or jan.lower() == "nan":
                continue

            title, price, error, _ = fetch_top_display_price(api_key, jan)

            if error:
                if "ãƒˆãƒ¼ã‚¯ãƒ³æ¯æ¸‡" in error:
                    log_box.insert(tk.END, f"ğŸª™ {i+1}/{total} {jan} â†’ ãƒˆãƒ¼ã‚¯ãƒ³æ¯æ¸‡ã€‚30åˆ†å¾…æ©Ÿã€‚\n")
                    log_box.see(tk.END)
                    time.sleep(300)#
                    continue
                elif "é€šä¿¡ã‚¨ãƒ©ãƒ¼" in error:
                    log_box.insert(tk.END, f"âš ï¸ {i+1}/{total} {jan} â†’ å‡¦ç†æ™‚é–“è¶…éã®ãŸã‚ã‚¹ã‚­ãƒƒãƒ—\n")
                    log_box.see(tk.END)
                    time.sleep(10)
                    continue

            results.append({
                "JANã‚³ãƒ¼ãƒ‰": jan,
                "ä¾¡æ ¼": price if price is not None else "Null",
                "å•†å“å": title or "",
                "å‚™è€ƒ": error or ""
            })

            # âœ… ãƒªã‚¢ãƒ«ã‚¿ã‚¤ãƒ å‡ºåŠ›ï¼ˆ1ä»¶ã”ã¨ï¼‰
            log_buffer.append(f"ğŸ• {i+1}/{total} ä»¶å‡¦ç†å®Œäº†\n")
            if len(log_buffer) >= 1:
                flush_logs(log_box, log_buffer)

            # âœ… 10ä»¶ã”ã¨ã«ä¿å­˜
            if (i + 1) % SAVE_INTERVAL == 0:
                pd.DataFrame(results).to_excel(output_file, index=False)
                log_box.insert(tk.END, f"ğŸ’¾ {i+1}ä»¶å®Œäº† â†’ ä¸€æ™‚ä¿å­˜ã—ã¾ã—ãŸã€‚\n")
                log_box.see(tk.END)

        # ğŸ”¹ æœ€å¾Œã«å†ä¿å­˜
        pd.DataFrame(results).to_excel(output_file, index=False)
        log_box.insert(tk.END, f"\nğŸ‰ å®Œäº†ï¼çµæœã‚’ã€Œ{output_file}ã€ã«ä¿å­˜ã—ã¾ã—ãŸã€‚\n")
        messagebox.showinfo("å®Œäº†", f"å‡¦ç†ãŒå®Œäº†ã—ã¾ã—ãŸï¼\nçµæœãƒ•ã‚¡ã‚¤ãƒ«: {output_file}")

    except Exception as e:
        log_box.insert(tk.END, f"âš ï¸ ã‚¨ãƒ©ãƒ¼ç™ºç”Ÿ: {e}\n{traceback.format_exc()}")
        messagebox.showerror("ã‚¨ãƒ©ãƒ¼", f"å‡¦ç†ä¸­ã«å•é¡ŒãŒç™ºç”Ÿã—ã¾ã—ãŸã€‚\n{output_file}")

    finally:
        start_button.config(state="normal")
        allow_sleep()

# ============================================================
# GUIæ§‹ç¯‰
# ============================================================
def create_gui():
    root = TkinterDnD.Tk()
    root.title("Keepaä¾¡æ ¼å–å¾—ãƒ„ãƒ¼ãƒ«")
    root.geometry("420x330")
    root.configure(bg="#f5f0e6")
    root.resizable(False, False)

    def on_close():
        if messagebox.askyesno("ç¢ºèª", "æœ¬å½“ã«çµ‚äº†ã—ã¾ã™ã‹ï¼Ÿ"):
            allow_sleep()
            root.destroy()
    root.protocol("WM_DELETE_WINDOW", on_close)

    var_topmost = tk.BooleanVar(value=True)
    chk_top = tk.Checkbutton(root, text="å¸¸ã«å‰é¢ã«è¡¨ç¤º", variable=var_topmost,
                             bg="#f5f0e6", font=("Meiryo", 9),
                             command=lambda: root.attributes("-topmost", var_topmost.get()))
    chk_top.pack(anchor="e", padx=10, pady=(3, 0))
    root.attributes("-topmost", True)

    tk.Label(root, text="Keepa APIã‚­ãƒ¼ï¼š", bg="#f5f0e6", font=("Meiryo", 10, "bold")).pack(anchor="w", padx=10, pady=2)
    api_entry = tk.Entry(root, width=55, show="*")
    api_entry.pack(padx=10)

    tk.Label(root, text="Excelãƒ•ã‚¡ã‚¤ãƒ«ï¼š", bg="#f5f0e6", font=("Meiryo", 10, "bold")).pack(anchor="w", padx=10, pady=2)
    file_label = tk.Label(root, text="ï¼ˆã“ã“ã«ãƒ‰ãƒ©ãƒƒã‚°ï¼†ãƒ‰ãƒ­ãƒƒãƒ— ã¾ãŸã¯ é¸æŠï¼‰", bg="white",
                          width=55, height=1, relief="groove", font=("Meiryo", 9))
    file_label.pack(padx=10, pady=2)

    def drop_file(event):
        filepath = event.data.strip("{}")
        file_label.config(text=filepath)
        file_label.filepath = filepath

    def select_file():
        filepath = filedialog.askopenfilename(filetypes=[("Excel files", "*.xlsx *.xls")])
        if filepath:
            file_label.config(text=filepath)
            file_label.filepath = filepath

    file_label.drop_target_register(DND_FILES)
    file_label.dnd_bind('<<Drop>>', drop_file)
    tk.Button(root, text="ãƒ•ã‚¡ã‚¤ãƒ«ã‚’é¸æŠ", command=select_file, font=("Meiryo", 9), width=18).pack(pady=3)

    frame_log = tk.Frame(root, bg="#f5f0e6")
    frame_log.pack(padx=10, pady=(5, 0), fill="both", expand=True)
    log_box = tk.Text(frame_log, height=6, width=55, font=("Meiryo", 8))
    log_box.pack(side="left", fill="both", expand=True)
    scrollbar = tk.Scrollbar(frame_log, command=log_box.yview)
    scrollbar.pack(side="right", fill="y")
    log_box.config(yscrollcommand=scrollbar.set)

    frame_buttons = tk.Frame(root, bg="#f5f0e6")
    frame_buttons.pack(pady=5)

    def force_stop():
        global STOP_FLAG
        STOP_FLAG = True
        log_box.insert(tk.END, "\nğŸ›‘ å¼·åˆ¶çµ‚äº†ãƒœã‚¿ãƒ³ãŒæŠ¼ã•ã‚Œã¾ã—ãŸã€‚\n")
        log_box.see(tk.END)

    start_button = tk.Button(frame_buttons, text="â–¶ é–‹å§‹", bg="#4CAF50", fg="white",
                             font=("Meiryo", 10, "bold"), width=14)
    start_button.pack(side="left", padx=15)

    tk.Button(frame_buttons, text="â–  å¼·åˆ¶çµ‚äº†", bg="#d9534f", fg="white",
              font=("Meiryo", 10, "bold"), width=14, command=force_stop).pack(side="right", padx=15)

    start_button.config(command=lambda: threading.Thread(
        target=start_process,
        args=(api_entry.get().strip(), getattr(file_label, "filepath", None), log_box, start_button),
        daemon=True
    ).start())

    root.mainloop()

# ============================================================
# ï¼˜æ—¥ã‚ˆã‚‹ã¾ã‚ã—ãŸã‚„ã¤
# ============================================================
if __name__ == "__main__":
    create_gui()
```

```python
import pandas as pd
import tkinter as tk
from tkinter import messagebox, scrolledtext
from tkinterdnd2 import TkinterDnD, DND_FILES
import os

def find_jan_column(df):
    """åˆ—åã‹ã‚‰JANã‚’è‡ªå‹•æ¤œå‡º"""
    for c in df.columns:
        if isinstance(c, str) and "JAN" in c.upper():
            return c
    return df.columns[0]

def determine_input_output(file1, file2):
    """ä¸­èº«ã‚’ç¢ºèªã—ã¦å…¥åŠ›ãƒ»å‡ºåŠ›ã‚’è‡ªå‹•åˆ¤å®šï¼ˆ1åˆ—ã®ã¿ã®æ–¹ã‚’å…¥åŠ›æ‰±ã„ï¼‰"""
    try:
        df1 = pd.read_excel(file1, header=None)
        df2 = pd.read_excel(file2, header=None)

        # âœ… 1åˆ—ã—ã‹ãªã„æ–¹ã‚’ã€Œå…¥åŠ›ãƒ•ã‚¡ã‚¤ãƒ«ã€ã¨åˆ¤å®š
        if df1.shape[1] == 1 and df2.shape[1] > 1:
            return file1, file2
        elif df2.shape[1] == 1 and df1.shape[1] > 1:
            return file2, file1
        elif df1.shape[1] == 1 and df2.shape[1] == 1:
            # ä¸¡æ–¹1åˆ—ã®å ´åˆ â†’ è¡Œæ•°ãŒçŸ­ã„æ–¹ã‚’å…¥åŠ›æ‰±ã„
            if len(df1) <= len(df2):
                return file1, file2
            else:
                return file2, file1
        else:
            raise ValueError("1åˆ—ã ã‘ã®Excelã‚’å…¥åŠ›ã¨ã—ã¦ãƒ‰ãƒ©ãƒƒã‚°ã—ã¦ãã ã•ã„ã€‚")
    except Exception as e:
        raise ValueError(f"ãƒ•ã‚¡ã‚¤ãƒ«åˆ¤å®šä¸­ã«ã‚¨ãƒ©ãƒ¼ãŒç™ºç”Ÿã—ã¾ã—ãŸ: {e}")

def process_files():
    """2ã¤ã®ãƒ•ã‚¡ã‚¤ãƒ«ãŒæƒã£ãŸæ™‚ç‚¹ã§å‡¦ç†"""
    if len(file_list) != 2:
        return  # 2ã¤æƒã†ã¾ã§å¾…æ©Ÿ

    try:
        input_path, output_path = determine_input_output(file_list[0], file_list[1])

        # ğŸŸ¢ å…¥åŠ›ï¼ˆãƒ˜ãƒƒãƒ€ãƒ¼ãªã—ï¼‰
        input_df = pd.read_excel(input_path, header=None)
        input_df.columns = ["JAN"]
        input_df["JAN"] = input_df["JAN"].astype(str).str.strip()

        # ğŸŸ¢ å‡ºåŠ›ï¼ˆãƒ˜ãƒƒãƒ€ãƒ¼ã‚ã‚Šï¼‰
        output_df = pd.read_excel(output_path)
        jan_col_output = find_jan_column(output_df)
        output_df[jan_col_output] = output_df[jan_col_output].astype(str).str.strip()

        # ğŸŸ¢ ä¸¦ã³æ›¿ãˆå‡¦ç†
        merged = pd.merge(input_df, output_df, left_on="JAN", right_on=jan_col_output, how="left")

        # âœ… å‡ºåŠ›å´ã®JANåˆ—ã‚’å‰Šé™¤
        merged.drop(columns=[jan_col_output], inplace=True)

        # ğŸ’¾ ä¿å­˜
        save_path = os.path.join(os.path.dirname(output_path), "JANæ•´åˆ—çµæœ.xlsx")
        merged.to_excel(save_path, index=False)

        messagebox.showinfo("å®Œäº†", f"JANæ•´åˆ—ãŒå®Œäº†ã—ã¾ã—ãŸï¼\n\nä¿å­˜å…ˆï¼š\n{save_path}")

    except Exception as e:
        messagebox.showerror("ã‚¨ãƒ©ãƒ¼", f"å‡¦ç†ä¸­ã«ã‚¨ãƒ©ãƒ¼ãŒç™ºç”Ÿã—ã¾ã—ãŸï¼š\n{e}")

def drop(event):
    """ãƒ‰ãƒ©ãƒƒã‚°ï¼†ãƒ‰ãƒ­ãƒƒãƒ—æ™‚ã®å‡¦ç†"""
    global file_list
    dropped_files = list(root.tk.splitlist(event.data))
    for f in dropped_files:
        if f not in file_list:
            file_list.append(f)

    # ãƒ•ã‚¡ã‚¤ãƒ«åã‚’è¡¨ç¤º
    file_text.delete(1.0, tk.END)
    file_text.insert(tk.END, "\n".join(os.path.basename(f) for f in file_list))

    # 2ã¤æƒã£ãŸã‚‰è‡ªå‹•å®Ÿè¡Œ
    if len(file_list) == 2:
        process_files()

def clear_files():
    """ãƒ‰ãƒ­ãƒƒãƒ—ã•ã‚ŒãŸãƒ•ã‚¡ã‚¤ãƒ«ã‚’å‰Šé™¤ï¼ˆãƒªã‚»ãƒƒãƒˆï¼‰"""
    global file_list
    file_list = []
    file_text.delete(1.0, tk.END)
    file_text.insert(tk.END, "ğŸ“‚ ãƒ•ã‚¡ã‚¤ãƒ«æœªé¸æŠ")
    messagebox.showinfo("ãƒªã‚»ãƒƒãƒˆ", "ãƒ•ã‚¡ã‚¤ãƒ«ãƒªã‚¹ãƒˆã‚’ãƒªã‚»ãƒƒãƒˆã—ã¾ã—ãŸã€‚")

# ğŸª¶ UIè¨­å®š
root = TkinterDnD.Tk()
root.title("JANæ•´åˆ—ãƒ„ãƒ¼ãƒ«")
root.geometry("520x370")
root.configure(bg="#f7f3e9")
root.resizable(False, False)  # ã‚µã‚¤ã‚ºå›ºå®š

title_label = tk.Label(
    root,
    text="JANæ•´åˆ—ãƒ„ãƒ¼ãƒ«\n\n"
         "â‘  å…¥åŠ›ï¼ˆ1åˆ—ã®Excelï¼‰ã¨å‡ºåŠ›ï¼ˆä¸€è¦§ã®Excelï¼‰ã‚’\n"
         "ã€€ã“ã“ã«ãƒ‰ãƒ©ãƒƒã‚°ï¼†ãƒ‰ãƒ­ãƒƒãƒ—ã—ã¦ãã ã•ã„ã€‚\n"
         "â‘¡ 2ã¤ã®ãƒ•ã‚¡ã‚¤ãƒ«ãŒæƒã†ã¨è‡ªå‹•ã§æ•´åˆ—ã—ã¾ã™ã€‚",
    bg="#f7f3e9",
    fg="#333",
    font=("Meiryo", 11)
)
title_label.pack(pady=10)

drop_frame = tk.Label(
    root,
    text="â¬‡ï¸ ã“ã“ã«2ã¤ã®ãƒ•ã‚¡ã‚¤ãƒ«ã‚’ãƒ‰ãƒ­ãƒƒãƒ— â¬‡ï¸",
    bg="#fffaf2",
    fg="#555",
    relief="ridge",
    width=50,
    height=5
)
drop_frame.pack(pady=5)

# ãƒ‰ãƒ­ãƒƒãƒ—ç™»éŒ²
drop_frame.drop_target_register(DND_FILES)
drop_frame.dnd_bind("<<Drop>>", drop)

# ğŸ“œ ãƒ•ã‚¡ã‚¤ãƒ«åè¡¨ç¤ºæ¬„ï¼ˆã‚¹ã‚¯ãƒ­ãƒ¼ãƒ«ä»˜ãï¼‰
file_text = scrolledtext.ScrolledText(
    root,
    width=55,
    height=4,
    bg="#fff",
    fg="#444",
    font=("Meiryo", 9),
    wrap="none"
)
file_text.insert(tk.END, "ğŸ“‚ ãƒ•ã‚¡ã‚¤ãƒ«æœªé¸æŠ")
file_text.pack(pady=5)

# ğŸ—‘ ãƒ•ã‚¡ã‚¤ãƒ«å‰Šé™¤ï¼ˆãƒªã‚»ãƒƒãƒˆï¼‰ãƒœã‚¿ãƒ³
clear_button = tk.Button(
    root,
    text="ğŸ—‘ ãƒ•ã‚¡ã‚¤ãƒ«ã‚’ãƒªã‚»ãƒƒãƒˆ",
    command=clear_files,
    bg="#f4cfa0",
    fg="#333",
    font=("Meiryo", 10, "bold"),
    relief="raised",
    width=20,
    height=1
)
clear_button.pack(pady=10)

file_list = []

root.mainloop()
```
