```python
import requests
import pandas as pd
import time

# ğŸŒ¸ Yahoo!ã‚·ãƒ§ãƒƒãƒ”ãƒ³ã‚° å•†å“æ¤œç´¢APIï¼ˆV3ï¼‰
API_URL = "https://shopping.yahooapis.jp/ShoppingWebService/V3/itemSearch"
APP_ID = "dj00aiZpPXlkOGd5bDlUcTlWRyZzPWNvbnN1bWVyc2VjcmV0Jng9MmE-"  # ã‚ãªãŸã®Client ID

# ğŸŒ· æ¤œç´¢è¨­å®š
SELLER_ID = "ishibashi"   # æ±æ€¥ãƒãƒ³ã‚ºYahoo!åº—
RESULTS_PER_CALL = 50      # 1å›ã§å–å¾—ã§ãã‚‹æœ€å¤§ä»¶æ•°
TOTAL_ITEMS = 2000         # ç›®æ¨™å–å¾—ä»¶æ•°
CALLS = TOTAL_ITEMS // RESULTS_PER_CALL

all_rows = []

print("ğŸ” Yahoo!ã‚·ãƒ§ãƒƒãƒ”ãƒ³ã‚°APIã‹ã‚‰å•†å“æƒ…å ±ã‚’å–å¾—ä¸­...")

for i in range(CALLS):
    start = i * RESULTS_PER_CALL + 1
    params = {
        "appid": APP_ID,
        "seller_id": SELLER_ID,
        "price_from": 100,
        "price_to": 100000,
        "results": RESULTS_PER_CALL,
        "start": start,
        "sort": "-score",      # ãŠã™ã™ã‚é †
        "in_stock": "true",
        "condition": "new"
    }

    try:
        response = requests.get(API_URL, params=params, timeout=10)
        data = response.json()
        hits = data.get("hits", [])
        if not hits:
            print(f"âš ï¸ {i+1}ãƒšãƒ¼ã‚¸ç›®ã«å•†å“ãªã—ã€‚")
            break

        for h in hits:
            jan = h.get("janCode") or ""
            price = h.get("price") or ""
            name = h.get("name") or ""
            url = h.get("url") or ""
            all_rows.append([jan, price, name, url])

        print(f"âœ… {i+1}ãƒšãƒ¼ã‚¸ç›®å®Œäº†ï¼ˆ{len(all_rows)}ä»¶ï¼‰")

        time.sleep(0.5)  # ğŸ’¡ APIè² è·è»½æ¸›ã®ãŸã‚å°‘ã—å¾…æ©Ÿ

    except Exception as e:
        print(f"âŒ ã‚¨ãƒ©ãƒ¼: {e}")
        time.sleep(5)

# ğŸŒ¼ DataFrameã«å¤‰æ›ã—ã¦Excelå‡ºåŠ›
df = pd.DataFrame(all_rows, columns=["JANã‚³ãƒ¼ãƒ‰", "ä¾¡æ ¼", "å•†å“å", "å•†å“URL"])
output_path = "handsnet_2000items.xlsx"
df.to_excel(output_path, index=False)

print("ğŸ‰ å®Œäº†ï¼Excelãƒ•ã‚¡ã‚¤ãƒ«ã‚’ä½œæˆã—ã¾ã—ãŸã€‚")
print(f"ğŸ“ ä¿å­˜å…ˆ: {output_path}")
```
